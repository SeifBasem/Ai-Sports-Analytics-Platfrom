```python
import cv2
from ultralytics import YOLO
from football_analysis import FootballAnalyzer
PLAYER_MODEL_PATH = r"D:\sports kitchen\best.pt"
FIELD_MODEL_PATH = r"D:\sports kitchen\yolo-football-pitch-detection.pt"
VIDEO_PATH = r"D:\sports kitchen\vids\mat.mp4"

OUTPUT_VIDEO_PATH = "output.mp4"
PLAYER_7_HEATMAP_PATH = "player_7_heatmap.png"
PLAYER_5_HEATMAP_PATH = "player_5_heatmap.png"
TEAM_0_HEATMAP_PATH = "team_0_heatmap.png"
TEAM_1_HEATMAP_PATH = "team_1_heatmap.png"
CSV_OUTPUT_PATH = "detections.csv"

model = YOLO(PLAYER_MODEL_PATH)
FIELD_DETECTION_MODEL = YOLO(FIELD_MODEL_PATH)
cap = cv2.VideoCapture(VIDEO_PATH)

if not cap.isOpened():
    raise ValueError(f"Could not open video: {VIDEO_PATH}")

fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

analyzer = FootballAnalyzer(
    model=model,
    fps=fps,
    device="cpu",
    field_model=FIELD_DETECTION_MODEL
)
analyzer.fit_team_classifier(VIDEO_PATH)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (width, height))
while True:
    ret, frame = cap.read()
    if not ret:
        break

    processed_frame = analyzer.process_frame(
        frame=frame,
        use_tiling=True,
    )

    out.write(processed_frame)

cap.release()
out.release()

analyzer.save_detections_csv(CSV_OUTPUT_PATH)
heatmap_img_player_7 = analyzer.generate_player_heatmap(
    player_id=7,
    player_name="messi"
)
cv2.imwrite(PLAYER_7_HEATMAP_PATH, heatmap_img_player_7)

heatmap_img_player_5 = analyzer.generate_player_heatmap(
    player_id=5,
    player_name="CR7"
)
cv2.imwrite(PLAYER_5_HEATMAP_PATH, heatmap_img_player_5)

team0_heatmap = analyzer.generate_team_heatmap(
    team_id=0,
    team_name="SAMS"
)
cv2.imwrite(TEAM_0_HEATMAP_PATH, team0_heatmap)

team1_heatmap = analyzer.generate_team_heatmap(
    team_id=1,
    team_name="AUC"
)
cv2.imwrite(TEAM_1_HEATMAP_PATH, team1_heatmap)

print("Done.")
print(f"Video saved to: {OUTPUT_VIDEO_PATH}")
print(f"CSV saved to: {CSV_OUTPUT_PATH}")
print(f"Player 7 heatmap saved to: {PLAYER_7_HEATMAP_PATH}")
print(f"Player 5 heatmap saved to: {PLAYER_5_HEATMAP_PATH}")
print(f"Team 0 heatmap saved to: {TEAM_0_HEATMAP_PATH}")
print(f"Team 1 heatmap saved to: {TEAM_1_HEATMAP_PATH}")
```
